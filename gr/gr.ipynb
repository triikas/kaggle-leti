```python
import matplotlib.pyplot as plt
x = [0.27,0.37,0.45,0.58,0.73,13.63,3.08,1.48,0.64,0.4,0.3]
y = list(map(lambda el : (el*10**(-3))/2, x))
print(y)
plt.plot(
    [1.8,2.1,2.4,2.7,3,4.1,4.5,6,7.5,9,10.5,], 
    y,
)

plt.xlabel('f')
plt.ylabel('Y(f)')

# plt.title('Зависимость от количества потоков')
plt.show()
```
