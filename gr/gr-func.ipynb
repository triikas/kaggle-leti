```python
import numpy as np
import matplotlib.pyplot as plt
import math

def func1(x):
    return 35.243*(10**(-15))-(35.243*(10**(-15)))*math.e**(-x/(10**(-6)))


def func2(x):
    return 35.243*(10**(-15))

x_values = np.linspace(0, 2*10**(-6), 1000)  # Задаем диапазон значений x

# Вычисляем значения функции для каждого x
y_values1 = [func1(x) for x in x_values]
y_values2 = [func2(x) for x in x_values]

# Строим график
plt.plot(x_values, y_values1, label='Выход')
plt.plot(x_values, y_values2, label='Вход')
plt.xlabel('t, c')
plt.ylabel('u(t), B')
# plt.title('Plot of Two Functions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Задание временного сигнала (например, синусоида)
fs = 1000  # Частота дискретизации, Hz
T = 2*10**(-6)    # Длительность, seconds
t = np.linspace(0, T, int(T * fs))
f = 50.0   # Частота сигнала, Hz
x = 0.5 * np.sin(2 * np.pi * f * t)

# Вычисление ДПФ
X = np.fft.fft(x)
N = len(X)
frequencies = np.fft.fftfreq(N, 1/fs)

# Вычисление амплитудного спектра
amplitude_spectrum = np.abs(X) / N

# Построение графика
plt.figure()
plt.plot(frequencies, amplitude_spectrum)
plt.title('Amplitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Определение функции
def f(x):
    # Избегаем деления на ноль
    return np.where(x == 0, 1, 2 * np.abs(np.sin(np.pi * x * 2 * 10**-6)) / (2 * np.pi * x))

# Определение диапазона x
x = np.linspace(-1e-5, 1e-5, 1000)

# Вычисление значений функции
y = f(x)

# Построение графика
plt.figure()
plt.plot(x, y)
plt.title(r'$f(x) = \frac{2 \cdot |\sin(\pi x \cdot 2 \cdot 10^{-6})|}{2 \pi x}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```
